```python
import matplotlib.pyplot as plt
import networkx as nx
import xgi
```

```python
H = xgi.read_hif("publications.hif.json")
# pos = xgi.pairwise_spring_layout(H, seed=0)
# H = xgi.largest_connected_hypergraph(H)
# pos_lcc = xgi.pairwise_spring_layout(H_lcc, seed=0)
```

```python
I = xgi.incidence_matrix(H, sparse=False)
plt.spy(I, aspect="auto")
plt.xlabel("Hyperedges")
plt.ylabel("Nodes")
plt.show()
```

```python
xgi.draw_bipartite(
    H,
    node_size=3,
    edge_marker_size=3,
    edge_marker_ec="lightblue",
    edge_marker_fc="navy",
    edge_marker_shape="o",
    dyad_color="grey",
    dyad_lw=0.5,
)
```

```python
A = xgi.adjacency_matrix(H, sparse=False)
plt.spy(A)
```

```python
G = xgi.Hypergraph([[i, j] for i, j in xgi.to_line_graph(H, s=1).edges])
xgi.draw(
    G, node_size=5, node_fc="navy", node_ec="lightblue", dyad_lw=0.5, dyad_color="grey"
)
```
